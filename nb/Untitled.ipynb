```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
xs = []
ys = []
colors = []
with open('perf_.txt', 'rU') as f:
    for line in f:
        s, x, y, _ = line.split()
        xs.append(int(x))
        ys.append(int(y))
        if s == 'original':
            colors.append(30)
        else:
            colors.append(10)
        
```

```python
plt.rcParams["figure.figsize"] =[12, 9]
plt.scatter(xs, ys, c=colors, alpha=0.5)
plt.show()
```

```python
plt.hist([y for i, y in enumerate(ys) if i%2==0], 20)
```
